### 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.¶

- Now using above information we will proceed further with that

```python
# Now we create a empty list where we will take url of the products
link=[]
for i in range(0,3):   #need to take data of 3 pages so running loop 3 times
    for j in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):   #this tag is of url of the product
        link.append(j.get_attribute('href'))
    try:
        driver.find_element_by_xpath("//li[@class='a-last']").click()     #This tag is to get to next page by click function
        time.sleep(5)
    except:
        print("No page further")      #if the pages of the product is less than 3 then this will print and will exit the loop
        break
```

```python
len(link)
```

```python
#creating empty list
Brand_Name=[]
Name_of_product=[]
Rating=[]
No_of_rating=[]
Price=[]
Return=[]
Expected_delivery=[]
Availability=[]
Other_Details=[]
```

```python
#now from each link we will extract the details from their xpath
for i in link:
    driver=webdriver.Chrome(r"C://Users/8.1/chromedriver")
    driver.get(i)
    time.sleep(2)
    try:
        Brand_Name.append(driver.find_element_by_xpath("//td[@class='a-span9']/span[1]").text)
    except NoSuchElementException:
        Brand_Name.append(driver.find_element_by_xpath("//a[@id='bylineInfo']").text)
    try:
        Name_of_product.append(driver.find_element_by_xpath("//h1[@id='title']").text)
    except NoSuchElementException:
        Name_of_product.append("-")
    try:
        Rating.append(driver.find_element_by_xpath("//span[@id='acrPopover']").get_attribute('title'))
    except NoSuchElementException:
        Rating.append("-")
        
    try:
        No_of_rating.append(driver.find_element_by_xpath("//a[@id='acrCustomerReviewLink']/span[1]").text)
    except NoSuchElementException:
        No_of_rating.append("-")
        
    try:
        Price.append(driver.find_element_by_xpath("//span[@id='priceblock_saleprice'] | //td[@class='a-span12']/span").text)
    except NoSuchElementException:
        Price.append("-")
        
    try:
        Return.append(driver.find_element_by_xpath("//div[@data-name='RETURNS_POLICY']//a[1]").text)
    except NoSuchElementException:
        Return.append("-")
    
    try:
        Expected_delivery.append(driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b[1]").text)
    except NoSuchElementException:
        Expected_delivery.append("-")
    
    try:
        Availability.append(driver.find_element_by_xpath("//div[@id='availability']/span[1]").text)
    except NoSuchElementException:
        Availability.append("-")
        
    try:
        Other_Details.append(driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']").text)
    except NoSuchElementException:
        Other_Details.append("-")
        
    driver.close()
```

```python
print(len(Brand_Name))
print(len(Name_of_product))
print(len(Rating))
print(len(No_of_rating))
print(len(Price))
print(len(Return))
print(len(Expected_delivery))
print(len(Availability))
print(len(Other_Details))
```

```python
#creating empty list
Brand_Name=[]
Smartphone_Name=[]
color=[]
RAM=[]
Storage=[]
Primary_Camera=[]
Secondary_Camera=[]
Display_Size=[]
Display_Resolution=[]
Processor=[]
Processor_Cores=[]
Battery_Capacity=[]
Price=[]
```

```python
#using the link and scraping the details we need 
for i in s_link:
    driver.get(i)
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    time.sleep(2)
    try:
        Brand_Name.append((driver.find_element_by_xpath("//span[@class='B_NuCI']").text).split(" ",1)[0])
    except NoSuchElementException:
        Brand_Name.append('-')
    try:
        Smartphone_Name.append(driver.find_element_by_xpath("//tr[@class='_1s_Smc row'][3]//li[@class='_21lJbe']").text)
    except NoSuchElementException:
        Smartphone_Name.append("-")
    try:
        color.append(driver.find_element_by_xpath("//tr[@class='_1s_Smc row'][4]//li[@class='_21lJbe']").text)
    except NoSuchElementException:
        color.append("-")
        
    try:
        RAM.append(driver.find_element_by_xpath("//div[@class='_3k-BhJ'][4]/table[@class='_14cfVK']//tr[2]//li[1]").text)
    except NoSuchElementException:
        RAM.append("-")
        
    try:
        Storage.append(driver.find_element_by_xpath("//div[@class='_3k-BhJ'][4]/table[@class='_14cfVK']//tr[1]//li[1]").text)
    except NoSuchElementException:
        Storage.append("-")
        
    try:
        Primary_Camera.append(driver.find_element_by_xpath("//div[@class='_3k-BhJ'][5]/table[@class='_14cfVK']//tr[2]//li[1]").text)
    except NoSuchElementException:
        Primary_Camera.append("-")
    
    try:
        Secondary_Camera.append(driver.find_element_by_xpath("//div[@class='_3k-BhJ'][5]/table[@class='_14cfVK']//tr[5]//li[1]").text)
    except NoSuchElementException:
        Secondary_Camera.append("-")
    
    try:
        Display_Size.append(driver.find_element_by_xpath("//div[@class='_3k-BhJ'][2]/table[@class='_14cfVK']//tr[1]//li[1]").text)
    except NoSuchElementException:
        Display_Size.append("-")
        
    try:
        Display_Resolution.append(driver.find_element_by_xpath("//div[@class='_3k-BhJ'][2]/table[@class='_14cfVK']//tr[2]//li[1]").text)
    except NoSuchElementException:
        Display_Resolution.append("-")
        
    try:
        Processor.append(driver.find_element_by_xpath("//div[@class='_3k-BhJ'][3]/table[@class='_14cfVK']//tr[2]//li[1]").text)
    except NoSuchElementException:
        Processor.append("-")
    
    try:
        Processor_Cores.append(driver.find_element_by_xpath("//div[@class='_3k-BhJ'][3]/table[@class='_14cfVK']//tr[3]//li[1]").text)
    except NoSuchElementException:
        Processor_Cores.append("-")

    try:
        Battery_Capacity.append(driver.find_element_by_xpath("//li[@class='_21Ahn-'][4]").text)
    except NoSuchElementException:
        Battery_Capacity.append("-")
    
    try:
        Price.append(driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']").text)
    except NoSuchElementException:
        Price.append("-")
    time.sleep(4)
```

```python
# making the dataframe
import pandas as pd
Phone=pd.DataFrame({})
Phone['Brand_Name']=Brand_Name
Phone['Smartphone_Name']=Smartphone_Name
Phone['color']=color
Phone['RAM']=RAM
Phone['Storage']=Storage
Phone['Primary_Camera']=Primary_Camera
Phone['Secondary_Camera']=Secondary_Camera
Phone['Display_Size']=Display_Size
Phone['Display_Resolution']=Display_Resolution
Phone['Processor']=Processor
Phone['Processor_Cores']=Processor_Cores
Phone['Battery_Capacity']=Battery_Capacity
Phone['Price']=Price
Phone['link']=s_link
```
