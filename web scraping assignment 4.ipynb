## 2. Scrape the details team India’s international fixtures from bcci.tv.

```python
driver=webdriver.Chrome(executable_path="C://Users/8.1/chromedriver")
```

```python
driver.get(x2)
```

```python
#creating empty list
Match_title=[]
Series=[]
Place=[]
Date=[]
Time=[]
```

```python
#Getting all the elements by there xpath 
for i in driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']"):  #for match title
    Match_title.append(i.text)
for i in driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__format']"):              #for series type
    Series.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']//span[2]"):      #for Time of play
    Time.append(i.text)
for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):      #forPlace where match will held 
    Place.append(i.text)
for i in driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']"):   #for date of match
    Date.append(i.text)
```

```python
#creating empty dataframe and then storing data in it
fixtures=pd.DataFrame({})
fixtures['Match_title']=Match_title
fixtures['Series']=Series
fixtures['Place']=Place
fixtures['Date']=Date
fixtures['Time']=Time
```

```python
#dataframe output
fixtures
```

```python
#now we need to clean the column Date, so we will replace '\n' with ',' 
for i in range(0,17):
    Date[i]=Date[i].replace("\n",',')
    Date[i]=' '.join(Date[i].split(',')[0:3])   #now we will slice the data to remove time and then join them
```

```python
#again assigning new data of the date
fixtures['Date']=Date
```

```python
#final dataframe
fixtures
```

## 10. Details of Datasets from UCI machine learning repositories. 
##### Url = https://archive.ics.uci.edu/
##### We have to find the following details:
##### A) Dataset name
##### B) Data type
##### C) Task
##### D) Attribute type
##### E) No of instances
##### F) No of attribute
##### G) Year

```python
#providing link to the driver
driver.get('https://archive.ics.uci.edu/ml/index.php')
```

```python
#now we will click on view all datasets
driver.find_element_by_xpath("//span[@class='whitetext'][2]//a[1]").click()
```

```python
#creating empty list
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
#As the total datasets are 559 so now we will run the loop starting from 2 because we are exculding header of table upto 563 as the total datasets are 559
for k in range(2,563):

#Here Tr tag is for rows and Td tag is for column
#getting all rows of data having dataset name column
    for i in driver.find_elements_by_xpath("//table[@border='1']//tr["+str(k)+"]/td[1]"):
        Dataset_name.append(i.text)
        
#getting all rows of data having datatype column
    for i in driver.find_elements_by_xpath("//table[@border='1']//tr["+str(k)+"]/td[2]"):
        Data_type.append(i.text) 
        
#getting all rows of data having Task column
    for i in driver.find_elements_by_xpath("//table[@border='1']//tr["+str(k)+"]//td[3]"):
        Task.append(i.text) 
        
#getting all rows of data having Attribute_type column
    for i in driver.find_elements_by_xpath("//table[@border='1']//tr["+str(k)+"]//td[4]"):
        Attribute_type.append(i.text) 
        
#getting all rows of data having No_of_instances column
    for i in driver.find_elements_by_xpath("//table[@border='1']//tr["+str(k)+"]//td[5]"):
        No_of_instances.append(i.text) 
        
#getting all rows of data having  No_of_attribute column
    for i in driver.find_elements_by_xpath("//table[@border='1']//tr["+str(k)+"]//td[6]"):
        No_of_attribute.append(i.text) 
        
#getting all rows of data having Year column
    for i in driver.find_elements_by_xpath("//table[@border='1']//tr["+str(k)+"]//td[7]"):
        Year.append(i.text) 

```

```python
#creating dataset
UCI=pd.DataFrame({})
UCI['Dataset_name']=Dataset_name
UCI['Data_type']=Data_type
UCI['Task']=Task
UCI['Attribute_type']=Attribute_type
UCI['No_of_instances']=No_of_instances
UCI['No_of_attribute']=No_of_attribute
UCI['Year']=Year
```

```python
#fianl dataset
UCI
```
